```python
# Import Tendorflow libraries and Keras
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
```

**Preprocess the Data**


*   Normalize pixel values to be between 0 and 1 for better model convergence
*   Reshape the data to match the input shape expected by the neural network model



```python
# Load and preprocess the datasets
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Reshape and normalize images
train_images = train_images.reshape((60000,28,28,1)).astype('float32') / 255
test_images = test_images.reshape((10000,28,28,1)).astype('float32') / 255

# One-hot encode the labels
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

**Build Neural Network Model**

* Create convolutional neural network (CNN) model using TensorFlow´s Keras API
* Compile the model with the Adam optimizer, sparse categorical crossentropy loss, and accuracy as the metrics


```python
# Build Convolutional Neural Network (CNN) Model
model = models.Sequential()#
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64,(3, 3), activation='relu'))
model.add(layers.MaxPooling2D(2, 2))
model.add(layers.Conv2D(64,(3, 3), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

**Train the Model**

* Train the model using the training data ('training_images' and 'train_labels') for 10 pochs
* Use the validation data ('test_images' and 'test_labels') to monitor the model´s performance during training

```python
# Train the model
model.fit(train_images, train_labels, epochs=10, batch_size=64, validation_split=0.1)
```

**Evaluate the Model**

Evaluate the model on the test data to assess its accuracy

```python
# Evaluate the model
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f'Test accuracy: {test_acc}')
```

**Make Predictions**

Use the train model to make predictions on the test data

```python
# Make Prediction
predictions = model.predict(test_images)
```

**Save and Load the Model**

```python
# Save the model
model.save('digit_classification_model.h5')

# Load the model
loaded_model = tf.keras.models.load_model('digit_classification_model.h5')
```
